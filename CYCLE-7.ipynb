1. Define a function which generates Fibonacci series up to n numbers.

```
# Normal method-1
n = int(input("Enter upto elements : "))
l = []
n1, n2 = 0, 1
nth = 1
print(n1)
l.append(n1)
if(n==1):
  print(nth)
else:
  while(nth <= n):
    print(nth)
    l.append(nth)
    nth = n1+n2
    n1 = n2
    n2 = nth


print(l)



    
```

```
Recursion
def rec(n):
  if(n<=1):
    return n
  else:
    return (rec(n-1) + rec(n-2))



n = int(input("Enter number : "))
print("Fibonacci sequence : ")
for i in range(n):
  print(rec(i))
```

```
# Normal method-2
n = int(input("Enter upto elements : "))
l = []
n1, n2 = 0, 1
nth = 0
count = 0
if(n==1):
  print(n1)
else:
  while(count < n):
    print(n1)
    l.append(n1)
    nth = n1+n2
    n1 = n2
    n2 = nth
    count += 1


print(l)



    
```

REVERESE FUCTION

```
def reverse(n):
  print(n)
  rev = 0
  while(n>0):
    r = n%10
    rev = (rev*10) + r
    n = n//10


  return rev



num = int(input("Enter number : "))
reverse(num)
```

ARMSTRONG NUMBER FUNCTION

 


```
def armg(n):
  temp = n
  order = len(str(n))
  sum = 0
  while(n > 0):
    r = n%10
    sum = sum + r**order
    n = n//10

  return sum



n = int(input("Enter number : "))
if(n == armg(n)):
  print(n," is an ARMSTRONG NUMBER")
else:
  print(n," Not an ARMSTRONG NUMBER")
```

```
#CALL_BY_VALUE
string = "RAVITEJA"

def call_by_value(string):
  string = "KARAVADI"
  print("Inside function = ",string)


call_by_value(string)
print("Outside string",string)
```

Implement a python script for Call-by-reference

```
#CALL_BY_REFERENCE

def add_more(my_list):
  my_list.append(50)
  print("inside = ",my_list)


my_list = [10,20,30,40,50]
my_list.append(60)
print("outside = ",my_list)
add_more(my_list)
```

Implement a python script for factorial of number by using recursion.

FACTORIAL #NORMAL-METHOD

```
def factorial(n):
  k = 1
  while(n>1):
    k = k*(n)
    n = n-1

  return k


n = int(input("Enter number : "))
factorial(n)
```

RECURSION


```
def fac(n):
  if(n==0):
    return n
  elif(n==1):
    return n
  else:
    return (n*fac(n-1))



n = int(input("Enter number : "))
print("factorial : ",fac(n))
```
