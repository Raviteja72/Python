1. LIST to TUPLE

```
n = int(input("Enter range of list : "))
k = []
print("Enter list elements : ")
for i in range(n):
  k.append(int(input()))




print(k)
print(tuple(k))
```

```
k = (1,2,3,4,5,6)
for i in k:
  print(i,end=" ")
  if(i == len(k)/2 ):
    break
print()
for i in k:
  if(i == len(k)/2):
    print(i+1,i+2,i+3)
```
