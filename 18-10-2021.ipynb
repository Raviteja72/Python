```python
# Perfect number 

a=int(input("Enter number : "))
s=0
for i in range(1,a):
    if(a%i == 0):
        s = s+i
        
if(s==a):
    print(a," is a perfect number ")
else:
    print(a," is not a perfect number ")
```

```python
# Strong number 

def fact(n):
    f=n
    while(n>1):
        f=f*(n-1)
        n=n-1
    return f


n=int(input("Enter number : "))
t=n
s=0
r=0
while(n!=0):
    r = n%10
    s = s+fact(r)
    n = n//10
    
    
if(t==s):
    print(t," is a Strong number ")
else:
    print(t," is not a strong number ")
```

```python
# Automorphic 

n=int(input("Enter number : "))

s=n*n

while(n>0):
    if(n%10 != s%10):
        print("Not Automorphic")
        break
    n=n//10
    s=s//10
    
else:
    print("Automorphic")
    
```

```python
# Harshad Number

n=int(input("Enter number : "))
k=n
s=0
while(k!=0):
    s=s+(k%10)
    k=k//10
print(s)
    
if(n%s == 0):
    print(n," is an Harshad number ")
else:
    print(n," is not an Harshad number ")
```

```python
# Abundant Number 

n=int(input("Enter number : "))
s=0
for i in range(1,n):
    if(n%i == 0):
        s = s+i
        
if(s>n):
    print(n," is an Abundant Number")
else:
    print(n," is not an Abundant Number")
```

```python
# Friendly Pair

a,b=map(int,input("Enter a and b values : ").split())
s1=0
s2=0
for i in range(1,a+1):
    if(a%i == 0):
        s1=s1+i
        

        
for i in range(1,b+1):
    if(b%i == 0):
        s2=s2+i
        
#print(s1,s2)

if(s1%a == s2%b):
    print(a,",",b," are Friendly Pair")
else:
    print(a,",",b," are Not Friendly Pair")
```

```python
# Sum of numbers at non-prime indices

l=list(map(int,input("Enter numbers : ").split()))
np=[0,1,]
n=len(l)
print("Length = ",n)
for i in range(2,n):
    for j in range(2,i):
        if(i%j == 0):
            np.append(i)
            break
            
            
            
s=0 
print(np)
for i in np:
    s = s+l[i]
    
    
print("Sum of non-primes : ",s)
```
